##Brands##

```python
from pyspark.sql.types import *
import pyspark.sql.functions as F
catalog_name = "ecommerce"
raw_data_file_path = f"/Volumes/{catalog_name}/raw/rawlanding/brands/*.csv"
#Let's create a schema for the DataFrame
brands_schema = StructType([
  StructField("brand_code", StringType(), False),
  StructField("brand_name", StringType(), True),
  StructField("category_code", StringType(), True)])

#Let's read the data from the file
brands_df = spark.read.csv(raw_data_file_path, header=True, schema=brands_schema, sep=",")
brands_df = brands_df.withColumn("_source_file", F.col("_metadata.file_path"))\
  .withColumn("_ingested_at", F.current_timestamp())
#Let's display the DataFrame
display(brands_df.limit(10))
brands_df.show(10)
# Let's write file in delta format
brands_df.write.format("delta").mode("overwrite").options(mergeSchema="true").saveAsTable(f"{catalog_name}.bronze.brz_brands")
```
